```python
# Cleaning the problems with datatypes, duplicated vaules in the initial dataframe
df_master = df_initial.copy()
```

```python
# Remove the duplicates
df_master.drop_duplicates('ListingKey', inplace = True)
```

```python
# Change to datetime datatype
df_master['ListingCreationDate'] = pd.to_datetime(df_master['ListingCreationDate'])
df_master['ClosedDate'] = pd.to_datetime(df_master['ClosedDate'])
df_master['LoanOriginationDate'] = pd.to_datetime(df_master['LoanOriginationDate'])
df_master['FirstRecordedCreditLine'] = pd.to_datetime(df_master['FirstRecordedCreditLine'])
df_master['DateCreditPulled'] = pd.to_datetime(df_master['DateCreditPulled'])
```

```python
# Change to Category data type
df_master['CreditGrade'] = df_master['CreditGrade'].astype('category')
df_master['LoanStatus'] = df_master['LoanStatus'].astype('category')
df_master['ProsperRating (numeric)'] = df_master['ProsperRating (numeric)'].astype('category')
df_master['ProsperRating (Alpha)'] = df_master['ProsperRating (Alpha)'].astype('category')
df_master['Occupation'] = df_master['Occupation'].astype('category')
df_master['EmploymentStatus'] = df_master['EmploymentStatus'].astype('category')
df_master['ListingCategory (numeric)'] = df_master['ListingCategory (numeric)'].astype('category')
df_master['IncomeRange'] = df_master['IncomeRange'].astype('category')
```

## Bivariate Exploration

> In this section, investigate relationships between pairs of variables in your
data. Make sure the variables that you cover here have been introduced in some
fashion in the previous section (univariate exploration).

```python
numerical_vars = ['BorrowerAPR', 'DebtToIncomeRatio', 'StatedMonthlyIncome', 'LoanOriginalAmount', 'MonthlyLoanPayment', 'Term'
               , 'ProsperScore', 'Investors', 'BorrowerRate', 'ListingCategory (numeric)']
```

```python
# Plotting correlation Plot
plt.figure(figsize=(10, 5))
sb.heatmap(df_master[numerical_vars].corr(), annot = True, fmt = '.3f', cmap = 'vlag_r', center = 0)
plt.show()
```

It can be observed that there a strong postive correlation between BorrowerAPR and BorrowerRate, and there is also
strong positive correlation between MonthlyLoanPayment and LoanOriginalAmount, which could possibly mean that, with an increase in Loan Amount there is an increase in monthly loan payments.

```python
# Plotting between Credit Grade and Loan Status
plt.figure(figsize=(15, 10))
ax = sb.countplot(data = df_master, x = 'CreditGrade', hue = 'LoanStatus')
ax.legend(loc = 1);
```

```python
# Plotting between Credit Grade and Income Range
plt.figure(figsize=(15, 10))
ax = sb.countplot(data = df_master, x = 'CreditGrade', hue = 'IncomeRange');
ax.legend(loc = 1);
```

```python
# Plotting between BorrowerAPR and BorrowerRate
sb.regplot(data = df_master, x = 'BorrowerAPR', y = 'BorrowerRate');
```

From the figure above we can observe that there is a positive correlation between BorrowerAPR and BorrowerRate.

```python
# Plotting Violin Plot for ProsperRating and BorrowerAPR
plt.figure(figsize=(10, 5))
sb.violinplot(data = df_master, x = 'ProsperRating (Alpha)',y = 'BorrowerAPR',color = base_color);
```

```python
plt.figure(figsize=(10, 5))
sb.violinplot(data = df_master, x = 'ProsperRating (Alpha)', y = 'BorrowerRate', color = base_color);
```

From the above two figures which are almost the same, we can say that BorroweRate and BorrowerAPR are strongly correlated with ProsperRating (Alpha).

```python
plt.figure(figsize=(10,5))
sb.regplot(data = df_master, y = 'BorrowerAPR', x = 'ProsperScore');
```

```python
plt.figure(figsize=(10,5))
sb.regplot(data = df_master, y = 'BorrowerRate', x = 'ProsperScore');
```

From the above two figures we can observe that BorrowerAPR and BorrowerRate are negatively correlated with the
ProsperScore.

```python
# Plot between LoanStatus and ProsperRating(Alpha)
plt.figure(figsize=(15, 8))
sb.countplot(data = df_master, x = 'LoanStatus', hue = 'ProsperRating (Alpha)', palette = 'viridis_r')
plt.xticks(rotation = 45)
plt.legend(loc = 1);
```

From the above plot we can observe that for prosper rating whose Loan Status is 'HR' is much more than that of Defaulted or Chargedoff, but it can also be seen that within their own category, Chargedoff and Defaulted have the highest number of 'HR' Prosper Ratings.

```python
# Plotting a violin plot between LoanStatus and BorrowerAPR
plt.figure(figsize=(15,8))
sb.violinplot(data = df_master, x = 'LoanStatus', y = 'BorrowerAPR', color = base_color);
plt.xticks(rotation = 45);
```

```python
plt.figure(figsize=(15,8))
sb.violinplot(data = df_master, x = 'LoanStatus', y = 'BorrowerRate', color = base_color);
plt.xticks(rotation = 45);
```

From the above two figures we can see that, BorrowerAPR and BorrowerRate are highly correlated with LoanStatus.

```python
cols = ['Term', 'BorrowerAPR', 'BorrowerRate', 'LoanOriginalAmount', 'LoanStatus']
g = sb.pairplot(df_master[cols], hue = 'LoanStatus');
```

From the pairplot we can observe that, it not that visually pleasing and confusing, so polishing is required.
But we can observe that BorrowerRate and BorrowerAPR are related.

```python
# Plot between Term, LoanStatus, and ProsperRating(Alpha)
sb.catplot(x = 'Term', hue = 'LoanStatus', col = 'ProsperRating (Alpha)', data = df_master, 
           kind = 'count', palette ='viridis_r', col_wrap = 3);
```

From the above figure, we can observer that most of the loans are of 36 months term and ProsperRating of A and AA have mostly been completed, while the HR ProsperRating has more number of Chargedoff and Cancelled combined Loans rather than completed.

```python
# Plot between LoanStatus, ProsperRating, BorrowerAPR
g = sb.FacetGrid(data = df_master, col = 'LoanStatus', col_wrap = 3, size = 4)
g.map(sb.pointplot, 'ProsperRating (Alpha)', 'BorrowerAPR', order = None, linestyles = "");
```

BorroweAPR is least for A and AA ProsperRaring, and is directly affecting the BorrowerAPR.

```python
# Plot between LoanStatus, ProsperRating, BorrowerAPR
g = sb.FacetGrid(data = df_master, col = 'LoanStatus', col_wrap = 3, size = 4)
g.map(sb.pointplot, 'ProsperRating (Alpha)', 'BorrowerRate', order = None, linestyles = "");
```

The relation between BorrowerAPR and BorrowerRate can be easily observed by comparing the previous two figures.

```python
# Plotting between LoanStatus, ProsperRating and LoanOriginalAmount
g = sb.FacetGrid(data = df_master, col = 'LoanStatus', col_wrap = 3, size = 4)
g.map(sb.pointplot, 'ProsperRating (Alpha)', 'LoanOriginalAmount', order = None, linestyles = "");
```

We can see that for ProsperRating (Alpha) HR has the LoanOriginalAmount of least in most of the figures, while
A and AA have fairly a large amount of LoanOriginalAmount.

```python
g = sb.FacetGrid(data = df_master, col = 'ProsperRating (Alpha)', col_wrap = 4, size = 3)
g.map(plt.scatter, 'BorrowerAPR', 'ProsperScore', alpha = 0.1)

plt.show()
```

ProsperRating (Alpha) for HR have higher BorroweAPR while AA have low BorroweAPR.
